```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#used for classification purpose
dataset=pd.read_csv("https://archive.ics.uci.edu/static/public/244/data.csv")
dataset
```

```python
features_names=['season',	'age',	'child_diseases',	'accident',	'surgical_intervention',	'high_fevers',	'alcohol',	'smoking',	'hrs_sitting',	'diagnosis']
features_names
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset['diagnosis'] = dataset['diagnosis'].replace({'N': 0, 'O': 1})
```

```python
X = dataset[['season',	'age',	'child_diseases',	'accident',	'surgical_intervention',	'high_fevers',	'alcohol',	'smoking',	'hrs_sitting',	'diagnosis']]
Y = dataset['diagnosis']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20)
```

```python
print(X_train.isnull().sum())
print(Y_train.isnull().sum())
```

```python
X_train = X_train.dropna()
```

```python
from sklearn.ensemble import RandomForestClassifier


forest = RandomForestClassifier(n_estimators=10)
forest.fit(X_train, Y_train)
```

```python
y_pred=forest.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test,y_pred) # because of the bootstraping dataset
#apart from the split
```
